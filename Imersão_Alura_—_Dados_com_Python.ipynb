```python
import pandas as pd

# Reload the DataFrame to ensure original column names
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv") # Recarrega o DataFrame para garantir nomes de coluna originais.

# Apply the correct renaming
df.rename(columns={
    "work_year": "ano",
    "experience_level": "senioridade",
    "employment_type": "tipo_emprego",
    "job_title": "cargo",
    "salary": "salario",
    "salary_currency": "moeda",
    "salary_in_usd": "salario_em_usd",
    "employee_residence": "residencia",
    "remote_ratio": "remoto",
    "company_location": "empresa",
    "company_size": "tamanho_empresa"
 },inplace=True) # Renomeia as colunas do DataFrame para português.

print(df.columns) # Exibe os novos nomes das colunas após a renomeação.
```

```python
df["remoto"].value_counts() # Conta a ocorrência de cada valor único na coluna 'remoto'.
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}) # Substitui os códigos de nível de experiência por descrições completas.

df['senioridade'].value_counts() # Exibe a contagem de valores na coluna 'senioridade' após a substituição.
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'L': 'Grande',
    'M': 'Médio',
    'S': 'Pequeno'
}) # Substitui os códigos de tamanho de empresa por descrições completas.

df['tamanho_empresa'].value_counts() # Exibe a contagem de valores na coluna 'tamanho_empresa' após a substituição.
```

```python
#criação de dataframe para exemplificar como lidar com dados nulos
import numpy as np #iniciei o codigo abaixo e depois a importação da biblioteca para lidar com numeros acima do df_salario
df_salarios = pd.DataFrame({
   "nome" : ["Ana", "Bruno", "Carlos", "Daniele", "Val"], #nomes ficticios
   "salario" : [4000, np.nan, 5000, np.nan, 1000000] #400 > Ana, np.nan (nulo) > Bruno, 500 > Carlos, np.nan (nulo) > Daniele
})
#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

'''Calcula mediana e substitui os nulos pela mediana
'''
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios["salario"].median())

df_salarios
```

```python
#quando o dados não faz sentido tirar a média
df_temperaturas = pd.DataFrame({
    "Dia" : ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura" : [30, np.nan, np.nan, 28, 27]
}) # Cria um DataFrame de exemplo para demonstrar preenchimento de dados.

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill() # Preenche os valores nulos na coluna 'Temperatura' com o último valor válido observado (forward-fill).
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia" : ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura" : [30, np.nan, np.nan, 28, 27]
}) # Cria um DataFrame de exemplo para demonstrar preenchimento de dados.

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill() # Preenche os valores nulos na coluna 'Temperatura' com o último valor válido observado (forward-fill).
df_temperaturas
#o bfill vem de back, usando o pandas
```

```python
#() para chamar {} para criar
df_cidades = pd.DataFrame({
    "nome" : ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade" : ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})

df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não informado")
df_cidades
```

```python
#Agora usando bibliotecas em conjunto
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='salario_em_usd')
plt.title("Salário médio por senioridade")
plt.xlabel("Senioridade") #rótulo
plt.ylabel("Salário médio anual (USD)")  #rótulo na lateral
plt.show() #apresentar imagem
```

```python
df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=True) #agrupar a média e ordenar do maior para o menor
```

```python
#Criando uma váriavel
ordem = df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=True).index #Mudando para True, altera a ordem também da linha de baixo

ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='salario_em_usd', order=ordem)
plt.title("Salário médio por senioridade")
plt.xlabel("Senioridade") #rótulo
plt.ylabel("Salário médio anual (USD)")  #rótulo na lateral
plt.show()
```

```python
#Agora outro tipo de gráfico para analisarmos a distribuição dos sálarios
plt.figure(figsize=(10,5))
sns.histplot(data=df_limpo['salario_em_usd'], bins=50, kde=True) #Histograma (histplot) + intervalos (bins) - se mudar o true para false, tira a linha do gráfico
plt.title("Distribuição dos salários anuais")
plt.xlabel("Salário em USB") #rótulo
plt.ylabel("Frequência")  #rótulo na lateral
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_limpo['salario_em_usd']) #Gráfico mais técnico
plt.title("Boxplot dos salários anuais")
plt.xlabel("Salário em USB")
plt.show
```

```python
#Váriavel
ordem_senioridade = ['Junior','Pleno','Senior', 'Executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade',y='salario_em_usd', data=df_limpo, order=ordem_senioridade)
plt.title("Boxplot da distribuição por senioridade")
plt.xlabel("Salário em USB")
plt.show()
```

```python
ordem_senioridade = ['Junior','Pleno','Senior', 'Executivo']

plt.figure(figsize=(8,5))
sns.boxplot(x='senioridade',y='salario_em_usd', data=df_limpo, order=ordem_senioridade, palette='Set2', hue='senioridade') #Usando paleta de cores
plt.title("Boxplot da distribuição por senioridade")
plt.xlabel("Salário em USB")
plt.show()
```

```python
import plotly.express as px
```

```python
#Código gerado pela IA atual (Tem diferença para a usada na aula online)
df_media_senioridade = df_limpo.groupby('senioridade')['salario_em_usd'].mean().reset_index()

# Usar a ordem definida anteriormente para consistência
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

fig = px.bar(
    df_media_senioridade,
    x='senioridade',
    y='salario_em_usd',
    title='Salário Médio Anual por Senioridade (Plotly)',
    labels={'senioridade': 'Senioridade', 'salario_em_usd': 'Salário Médio Anual (USD)'},
    category_orders={'senioridade': ordem_senioridade},
    color='senioridade', # Adiciona cores diferentes para cada barra
    color_discrete_sequence=px.colors.qualitative.Set2 # Usa uma paleta de cores agradável
)

fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray':ordem_senioridade})
fig.show()
```

```python
#Código gerado pela IA da aula
senioridade_media_salario = df_limpo.groupby('senioridade')['salario_em_usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    senioridade_media_salario,
    x='senioridade',
    y='salario_em_usd', # Corrigido de 'usd' para 'salario_em_usd'
    title='Média Salarial por Senioridade',
    labels={'senioridade': 'Nível de Senioridade', 'salario_em_usd': 'Média Salarial Anual (USB)'}
)

fig.show()
```
